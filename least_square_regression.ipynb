```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis] #utk buat axis baru tapi ga keliatan

# rumus = (A.T x A)^-1 . (A.T x Y)

# first = np.linalg.inv(np.dot(A.T, A))
# second = np.dot(first, A.T)
# result = np.dot(second, y)

result = np.dot(np.linalg.pinv(A), y) #shortcut
print(result)


plt.plot(x, y, 'co') # b -> blue; o -> bulet kalo mau line itu pake - gantiin o
#y = mx+b
#m -> result 0
#b -> result 1
plt.plot(x, result[0] * x + result[1], 'r-')
plt.title("Linear Square Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
